Crie uma lista de valores inteiros randômicos e obtenha em uma função com retorno múltiplo os valores: máximo, mínimo e a média.

_Hint_: use a função random.sample(range(<número_máximo), <números>)

```python
import random

lista_rand = random.sample(range(0,10000), 10000)
```

```python
lista_rand
```

```python
def min_max_media(lista):
    min = lista[0]
    max = lista[1]
    soma = 0
    
    for item in lista:
        soma += item
        if item > max:
            max = item
        if item < min:
            min = item
    
    media = soma/len(lista)
    
    return min, max, media
```

```python
min_max_media(lista_rand)
```

```python
def min_max_media_built_in(lista):
    return min(lista), max(lista), sum(lista) / len(lista)
```

```python
min_max_media_built_in(lista_rand)
```
