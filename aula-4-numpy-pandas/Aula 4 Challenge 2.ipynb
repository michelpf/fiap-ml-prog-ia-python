Utilize o arquivo "iris.data.txt" que contem os dados do projeto Íris, em partilar o tamanho das sépalas, para normalizar seus dados, de tal forma que os valores fiquem entre 0 e 1.

```python
import numpy as np
```

```python
sepal_txt_file = np.loadtxt("iris.data.txt", delimiter=',', dtype='float', usecols=[0])
sepal_txt_file
```

```python
sepal_max, sepal_min = sepal_txt_file.max(), sepal_txt_file.min()
sepal_norm = (sepal_txt_file - sepal_min)/(sepal_max - sepal_min)
sepal_norm
```
