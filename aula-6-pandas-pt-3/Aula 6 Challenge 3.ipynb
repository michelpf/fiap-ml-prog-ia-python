Utilizando o dataframe ajustado da challenge 1, padronize seus dados para exibir valores numéricos ao invés de dados categorizados no atributo de classe, cidade e estado.

Exemplo, ao invés de true ou false, utilizar 1 ou 0.

```python
import numpy as np
import pandas as pd
```

```python
df_pimas = pd.read_csv("data/pima-data-norm-city.csv", index_col=0)
df_pimas.head(5)
```

```python
class_map = {True: 1, False: 0}
city_map = {"Agra": 1, "Bangalore": 2, "Hyderabad": 3, "Lucknow": 4, "Mumbai": 5, "Patna": 6}
state_map = {"Bihar": 1, "Kamataka": 2, "Maharashtra": 3, "Telangana": 4, "Uttar Pradesh": 5}

df_pimas["diabetes"] = df_pimas["diabetes"].map(class_map)
df_pimas["city"] = df_pimas["city"].map(city_map)
df_pimas["state"] = df_pimas["state"].map(state_map)
```
