```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_stock = pd.read_excel("data/data_akbilgic.xlsx")
df_stock.head(5)
```

```python
df_stock_pivoted = df_stock.pivot(index="day", columns="month_year", values="BOVESPA")
df_stock_pivoted.head(5)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 8)
sns.heatmap(df_stock_pivoted)
```
